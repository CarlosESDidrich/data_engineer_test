# Tratamento dos dados VRA
    Durante o Tratamento foi utilizado a def de Felipe Avelar, disponivel no stackOverflow, para realizar a normalização do cabeçalho, devido o padrão das strings serem Camel case. Nesse tratamento optei por deixar somente uma cell para agilizar o processo de tratamento, todos foram tratados e exportados como json, para a pasta VRA_Tratados.
    
## Desafios abordados   
- Carregar os dados de VRA
  - Normalizar o cabeçalho para snake case
  - Salvar estes dados

```python
# Bibliotecas usadas
import pandas as pd
import os
import ast
import json
import re
```

## Def de normalização do cabeçalho

```python
def snake_case(valor):
    valor = re.findall('[A-Z]*[^A-Z]*', valor) # dá um split nas letras maiúsculas
    if valor[0] == '_':
        valor[0] = ''
    valor = [x.lower() for x in valor]
    return '_'.join(valor)[:-1]
```

## Tratamento dados 

```python
# Carregando os dados .json
caminho_arquivo = 'C:/Users/Didrich/Documents/Python Scripts/DataEnginnerTest/VRA/VRA_Brutos/VRA_20218.json'  

if not os.path.isfile(caminho_arquivo):
    exit(1)  

with open(caminho_arquivo, 'r', encoding='utf-8-sig') as arquivo:
    conteudo_arquivo = arquivo.read()
    df = json.loads(conteudo_arquivo)
    df = pd.DataFrame(df)
    
# Aplicar a função a todas as colunas do DataFrame
df.columns = [snake_case(col) for col in df.columns]

# Verificar o resultado
df.head()
```

```python
# Exportar o DataFrame para um arquivo JSON
import os

# Especifique o caminho completo para a pasta onde deseja salvar o arquivo JSON
pasta_especifica = 'C:/Users/Didrich/Documents/Python Scripts/DataEnginnerTest/VRA/VRA_tratados'

# Certifique-se de que a pasta existe, caso contrário, crie-a
if not os.path.exists(pasta_especifica):
    os.makedirs(pasta_especifica)

    caminho_arquivo_json = os.path.join(pasta_especifica, 'VRA_20218_Teste.json')

# Exporte o DataFrame para o arquivo JSON
df.to_json(caminho_arquivo_json, orient='records', lines=True)

print(f'O DataFrame foi exportado para {caminho_arquivo_json}')
```
